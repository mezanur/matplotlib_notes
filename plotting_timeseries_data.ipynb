```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
df = pd.read_csv('data/covid19Bangladesh.csv', parse_dates=['Date'], index_col='Date')
```

```python
df.drop(columns=['Day'], inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot_date(df.index, df['Total Deaths'], marker='.', linestyle='-', color='#ff304f')
#plt.plot_date(df.index, df['Total Cases'], marker='.', linestyle='-', color='#085f63')
#plt.yscale('log')
#plt.gcf().autofmt_xdate()
date_formatter = mpl_dates.DateFormatter('%b %d')
plt.gca().xaxis.set_major_formatter(date_formatter)
plt.title('Total Deaths')
plt.xlabel('Dates')
plt.tight_layout()
plt.show()
```
