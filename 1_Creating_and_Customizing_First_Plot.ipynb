```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
```

```python
df.dropna(how='any', subset=['ConvertedComp', 'Age', 'LanguageWorkedWith'], inplace=True)
df['Age'] = df['Age'].astype(int)
```

```python
age_grp = df.groupby('Age')
```

```python
dev_median = age_grp['ConvertedComp'].median().loc[18:60]
ages_x = dev_median.index
dev_y = dev_median.values
```

```python
print(dev_median.shape, py_y.shape, js_y.shape)
```

```python
js_df = df[df['LanguageWorkedWith'].str.contains('JavaScript', na=False)]
js_dev_median = js_df.groupby('Age')['ConvertedComp'].median().loc[18:60]
js_y = js_dev_median.values
```

```python
py_df = df[df['LanguageWorkedWith'].str.contains('Python', na=False)]
py_dev_median = py_df.groupby('Age')['ConvertedComp'].median().loc[18:60]
py_y = py_dev_median.values
```

```python
plt.plot(ages_x, js_y, color='#28c7fa', label='JavaScript')
plt.plot(ages_x, py_y, color='#f26d5b', label='Python')
plt.plot(ages_x, dev_y, linewidth=1, color='#0b8457', linestyle='-', label='All Devs')
#plt.grid(True)
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Salary(USD)')
plt.legend()
#plt.savefig('savefig/med_sal.svg')
#plt.savefig('med_sal.png', dpi=800)
plt.tight_layout()
plt.show()
```
