```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens' , edgecolors='k', linewidths=1, alpha=1)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()
plt.show()
```

```python
yt_data = pd.read_csv('data/yt_data.csv')
view = yt_data['view_count']
likes = yt_data['likes']
ratios = yt_data['ratio']
```

```python
plt.scatter(view, likes, edgecolors='k', linewidths=1, alpha=.60, c=ratios, cmap='summer')
cbar = plt.colorbar()
cbar.set_label('Like:Dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```
