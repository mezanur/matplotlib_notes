```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
plt.style.use('seaborn')
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('data/survey_results_schema.csv', index_col='Column')
```

```python
webframe = df['WebframeWorkedWith'].dropna()
```

```python
counter = Counter()
```

```python
for x in webframe.str.split(';'):
    counter.update(x)
```

```python
sorted_count = counter.most_common()
```

```python
wf = []
popularity = []
for key, value in sorted_count:
    wf.append(key)
    popularity.append(value)
```

```python
plt.barh(wf, popularity, color='#3d5af1')
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.tick_params(labelright=True, labelleft=False)
plt.title('Most Popular Web Framework')
plt.show()
```
