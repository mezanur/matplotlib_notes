```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
```

```python
country_grp = df.groupby('Country')
```

```python
countries = list(country_grp['ConvertedComp'].median().sort_values(ascending=False).iloc[0:20].index)
pay = list(country_grp['ConvertedComp'].median().sort_values(ascending=False).iloc[0:20].values)
```

```python
countries[countries.index("Lao People's Democratic Republic")] = 'Laos'
countries[countries.index('Hong Kong (S.A.R.)')] = 'Hong Kong'
```

```python
plt.barh(countries, pay, color='#0e153a', height=.5)
plt.title('Countries with highest median developer salary'.title(), size=10)
plt.xlabel('Median yearly salary (USD)', size=10)
plt.tick_params(axis='y', labelsize=8)
plt.xticks(fontsize=8)
plt.tight_layout()
#plt.savefig('savefig/countries_with_highest_dev_sal.png', dpi=1200)
plt.show()
```
