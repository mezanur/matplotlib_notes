## Two Ways of Making Plots
1. Object Oriented Style

```python
x = np.linspace(0,2,100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('simple plot')
ax.legend()
plt.show()
```

2. Pyplot Style

```python
x = np.linspace(0,2,100)
plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadratic')
plt.plot(x, x**3, label='Cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

```python
def my_plotter(ax, datax, datay, param_dict):
    out = ax.plot(datax, datay, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4,100)
```

```python
fig, ax = plt.subplots()
my_plotter(ax, data1, data2, {'color':'g', 'marker':'o'})
```

```python
fig, axs = plt.subplots(1,2)
ax1, ax2 = axs
my_plotter(ax1, data1, data2, {'color':'g', 'marker':'H'})
ax1.set_xlabel('xlabel 1')
ax1.set_ylabel('ylabel 1')
my_plotter(ax2, data3, data4, {'color':'k', 'marker':'o'})
ax2.set_xlabel('xlabel 2')
ax2.set_ylabel('ylabel 2')
plt.tight_layout()
```
