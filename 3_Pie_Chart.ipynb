```python
import matplotlib.pyplot as plt
from collections import Counter
import csv
```

```python
plt.style.use('seaborn')
```

```python
with open('data/survey_results_public.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    counter = Counter()
    for row in csv_reader:
        counter.update(row['LanguageWorkedWith'].split(';'))
```

```python
slices = [tup[1] for tup in counter.most_common(5)]
labels = [tup[0] for tup in counter.most_common(5)]
colors = ['#324e7b', '#86a6df', '#5068a9', '#324e7b', '#86a6df']
colors2 = ['#e8ffe8', '#a6fff2', '#74f9ff', '#00e0ff']
explode = [0, 0, 0, 0.1, 0]
```

```python
def give_pct(pct):
    return str(round(pct,2))+'%' +' ['+str(round(pct*sum(slices)/100))+']'
```

```python
plt.pie(slices, labels=labels, colors=colors, startangle=90,
        explode=explode, #shadow=True,
        autopct=give_pct, pctdistance=.7, wedgeprops={'edgecolor':'black', 'linewidth':'.0'},
       counterclock=False, textprops={'color':'black', 'backgroundcolor':'white'})
plt.title('Most Common Programming Languages')
plt.gca().add_artist(plt.Circle((0,0), .4, color='w'))
plt.show()
```
