```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
plt.style.use('ggplot')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y, linestyle='-')
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```

```python
btc_df = pd.read_csv('data/btc_datetime.csv', parse_dates=['Date'])
```

```python
btc_df
```

```python
price_date = btc_df['Date']
price_close = btc_df['Close']
```

```python
plt.figure(figsize=(10,6))
plt.plot_date(price_date, price_close, marker='.', linestyle='-', color='b')
#plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b %d')
plt.gca().xaxis.set_major_formatter(date_format)
plt.title('Bitcoin Price')
plt.ylabel('Price in USD')
plt.xlabel('Dates')
plt.show()
```
