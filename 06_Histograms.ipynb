```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-paper')
```

```python
df = pd.read_csv('data/survey_results_public.csv')
```

```python
ages = df['Age'].dropna().astype(int)
```

```python
median_age = df['Age'].median()
```

```python
bins = np.linspace(10,100,10, dtype='int')
```

```python
fig, (ax, axlog) = plt.subplots(2,1)

ax.set_title('Ages of Respondent', size=10)
ax.hist(ages, bins=bins, color='#cabbe9', edgecolor='#071a52')
ax.axvline(median_age, color='k', linewidth=1, label='Median Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Total Respondents')
ax.legend(shadow=True)

#With Log Scale
axlog.set_title('Ages of Respondent(Logaritmic Scale)', size=10)
axlog.hist(ages, bins=bins, log=True, color='#a6acec', edgecolor='#071a52')
axlog.axvline(median_age, color='k', linewidth=1, label='Median Age')
axlog.set_xlabel('Ages')
axlog.set_ylabel('Total Respondents')
axlog.minorticks_off()
axlog.legend(shadow=True)
fig.tight_layout()
#fig.savefig('data/histogram.svg')
```
