```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from collections import Counter
```

```python
with open('data/survey_results_public.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    counter = Counter()
    for row in csv_reader:
        counter.update(row['LanguageWorkedWith'].split(';'))
```

```python
langs = [tup[0] for tup in counter.most_common(15)]
pop = [tup[1] for tup in counter.most_common(15)]
```

```python
plt.barh(langs, pop)
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```
